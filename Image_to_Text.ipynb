```python
#import the pytesseract 
import pytesseract

#import the Image from Pillow
from PIL import Image

#Opening of the Image
img = Image.open('temp.png')

#Displaying the image 'temp.png'
display(Image.open("temp.png"))

#image_to_string() is the function in pytesseract with converts the text
#in the image to text and returns the text as string type
text = pytesseract.image_to_string(img)

print("The Text in the Image is : ")

#Printing the text in the image
print(text)
```
